#  Task 3: Data Visualization

This section visualizes student performance data to uncover patterns and support decision-making using Matplotlib and Seaborn.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/kaggle/input/student-performance-factors/StudentPerformanceFactors.csv')
```

## 1. Distribution of Exam Scores

Understand the spread and central tendency of students' exam performance.


```python
plt.figure(figsize=(8,5))
sns.histplot(df['Exam_Score'], kde=True, color='skyblue')
plt.title('Distribution of Exam Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

## 2. Correlation Heatmap

Examine how numerical variables correlate with each other.


```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## 3. Hours Studied vs Exam Score

Explore the relationship between study hours and performance.


```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Hours_Studied', y='Exam_Score', hue='Gender')
plt.title('Hours Studied vs Exam Score')
plt.show()
```

## 4. Parental Education vs Exam Score

Compare exam results across different parental education levels.


```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Parental_Education_Level', y='Exam_Score', palette='pastel')
plt.xticks(rotation=45)
plt.title('Parental Education Level vs Exam Score')
plt.show()
```

## 5. Distribution of Categorical Features

Visualize the counts for important categorical variables.


```python
categorical_cols = ['Parental_Involvement', 'Internet_Access', 'School_Type', 'Tutoring_Received']
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, palette='Set2')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

## 6. Sleep Hours vs Exam Score

Visualize how sleep duration impacts exam scores.


```python
df['Sleep_Bin'] = pd.cut(df['Sleep_Hours'], bins=[0, 4, 6, 8, 10], labels=['<4','4-6','6-8','8+'])
plt.figure(figsize=(8,5))
sns.boxplot(x='Sleep_Bin', y='Exam_Score', data=df)
plt.title('Sleep Hours vs Exam Score')
plt.show()
```
